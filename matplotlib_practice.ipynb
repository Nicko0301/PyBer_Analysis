```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth= 2, label = 'Boston')
# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = "^", color = "green", linewidth = 2, label = 'Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0, 45)
ax.set_title("Pyber Fare by Month")
ax.grid()
ax.legend()
```

```python
plt.bar(x_axis, y_axis, color = 'green', label = 'Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.legend
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color = 'magenta', label = 'Boston')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('Pyber Fare by Month')
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Bubble plot
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)
    
y_axis_larger
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Create Scatter plot using object oriented approach
fig, ax = plt.subplots(figsize = (8,8))
ax.scatter(y_axis, x_axis, c = "skyblue", s = [i*5 for i in y_axis], alpha = .2, linewidth = 2, edgecolors = 'black', label = 'Boston')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.set_ylim('Dec', 'Jan')
ax.set_xlim(0,50)
ax.legend()
```

```python
# Create Pie Chart
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create pie chart using object oriented approach.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)


fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, 
       labels = x_axis, 
       colors = colors, 
       explode = explode_values,
      autopct = '%.1f%%',
      shadow = True,
      startangle=95,
      counterclock = False)
plt.show()
```
